```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL of the website we want to scrape
url = "http://books.toscrape.com/"

# Send a GET request to the webpage
response = requests.get(url)

# Check if the request was successful
print(response.status_code)  # 200 means success

# Store the page content
webpage = response.text
```

```python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(webpage, "html.parser")

# Print the first 500 characters to see the structure
print(soup.prettify()[:500])
```

```python
# Find all book titles
books = soup.find_all("h3")  # 'h3' contains book titles
titles = [book.a["title"] for book in books]  # extract title attribute

# Check the first 10 titles in
print(titles[:10])
```

```python
# Convert to DataFrame
df = pd.DataFrame(titles, columns=["Book Title"])

# Save to CSV
df.to_csv("books.csv", index=False)
```
