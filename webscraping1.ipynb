```python
import requests
from bs4 import BeautifulSoup


```

```python
# Step 1: Target website
url = "https://httpbin.org/html"

# Step 2: Fetch the page
response = requests.get(url)
html = response.text

```

```python
# Step 3: Parse HTML
soup = BeautifulSoup(html, "html.parser")

# Extract H1 safely
h1_tag = soup.find("h1")
h1 = h1_tag.text if h1_tag else "No H1 found"

print("H1 Heading:", h1)
```
